```python
import requests
```

```python
urls = {
    'adult_obesity': ('https://ghoapi.azureedge.net/api/NCD_BMI_30C', 'Adult'),
    'child_obesity': ('https://ghoapi.azureedge.net/api/NCD_BMI_PLUS2C', 'Child/Adolescent'),
    'adult_malnutrition': ('https://ghoapi.azureedge.net/api/NCD_BMI_18C', 'Adult'),
    'child_malnutrition': ('https://ghoapi.azureedge.net/api/NCD_BMI_MINUS2C', 'Child/Adolescent')
}
```

```python
import pandas as pd

# 1. Load adult obesity data
url1 = 'https://ghoapi.azureedge.net/api/NCD_BMI_30C'
response1 = requests.get(url1)
data1 = response1.json()['value']
df_adult_obesity = pd.DataFrame(data1)
df_adult_obesity['age_group'] = 'Adult'

```

```python
# 2. Load child obesity data
url2 = 'https://ghoapi.azureedge.net/api/NCD_BMI_PLUS2C'
response2 = requests.get(url2)
data2 = response2.json()['value']
df_child_obesity = pd.DataFrame(data2)
df_child_obesity['age_group'] = 'Child'
```

```python
# 3. Load adult malnutrition data
url3 = 'https://ghoapi.azureedge.net/api/NCD_BMI_18C'
response3 = requests.get(url3)
data3 = response3.json()['value']
df_adult_malnutrition = pd.DataFrame(data3)
df_adult_malnutrition['age_group'] = 'Adult'
```

```python
# 4. Load child malnutrition data
url4 = 'https://ghoapi.azureedge.net/api/NCD_BMI_MINUS2C'
response4 = requests.get(url4)
data4 = response4.json()['value']
df_child_malnutrition = pd.DataFrame(data4)
df_child_malnutrition['age_group'] = 'Child'
```

```python
# 5. Combine obesity datasets
df_obesity = pd.concat([df_adult_obesity, df_child_obesity], ignore_index=True)
```

```python
# 6. Combine malnutrition datasets
df_malnutrition = pd.concat([df_adult_malnutrition, df_child_malnutrition], ignore_index=True)
```

```python
# 7. Filter years
df_obesity = df_obesity[(df_obesity['TimeDim'] >= 2012) & (df_obesity['TimeDim'] <= 2022)]
df_malnutrition = df_malnutrition[(df_malnutrition['TimeDim'] >= 2012) & (df_malnutrition['TimeDim'] <= 2022)]
```

```python
# Data Cleaning
```

```python
columns_needed = ['TimeDim', 'Dim1', 'NumericValue', 'Low', 'High', 'ParentLocation', 'SpatialDim', 'age_group']

df_obesity = df_obesity[columns_needed].copy()
df_malnutrition = df_malnutrition[columns_needed].copy()
```

```python
# Renaming the columns
rename_columns = {
    'TimeDim': 'Year',
    'Dim1': 'Gender',
    'NumericValue': 'Mean_Estimate',
    'Low': 'LowerBound',
    'High': 'UpperBound',
    'ParentLocation': 'Region',
    'SpatialDim': 'Country'
}

df_obesity.rename(columns=rename_columns, inplace=True)
df_malnutrition.rename(columns=rename_columns, inplace=True)
```

```python
df_obesity['Gender'] = df_obesity['Gender'].replace({
    'SEX_MLE': 'Male',
    'SEX_FMLE': 'Female',
    'SEX_BTSX': 'Both'
})

df_malnutrition['Gender'] = df_malnutrition['Gender'].replace({
    'SEX_MLE': 'Male',
    'SEX_FMLE': 'Female',
    'SEX_BTSX': 'Both'
})
```

```python
import pycountry
```

```python
def convert_country(code):
    try:
        return pycountry.countries.get(alpha_3=code).name
    except:
        return code

df_obesity['Country'] = df_obesity['Country'].apply(convert_country)
df_malnutrition['Country'] = df_malnutrition['Country'].apply(convert_country)
```

```python
special_cases = {
    'GLOBAL': 'Global',
    'WB_LMI': 'Low & Middle Income',
    'WB_HI': 'High Income',
    'WB_LI': 'Low Income',
    'EMR': 'Eastern Mediterranean Region',
    'EUR': 'Europe',
    'AFR': 'Africa',
    'SEAR': 'South-East Asia Region',
    'WPR': 'Western Pacific Region',
    'AMR': 'Americas Region',
    'WB_UMI': 'Upper Middle Income'
}

df_obesity['Country'] = df_obesity['Country'].replace(special_cases)
df_malnutrition['Country'] = df_malnutrition['Country'].replace(special_cases)
```

```python
df_obesity
```

```python
# adding new column CI Width
df_obesity['CI_width'] = df_obesity['UpperBound'] - df_obesity['LowerBound']
df_malnutrition['CI_width'] = df_malnutrition['UpperBound'] - df_malnutrition['LowerBound']
```

```python
# adding new column obesity_level

def get_obesity_level(value):
    if value >= 30:
        return 'High'
    elif value >= 25:
        return 'Moderate'
    else:
        return 'Low'

df_obesity['Obesity_Level'] = df_obesity['Mean_Estimate'].apply(get_obesity_level)
```

```python
# adding new column Malnutrition_Level

def get_malnutrition_level(value):
    if value >= 20:
        return 'High'
    elif value >= 10:
        return 'Moderate'
    else:
        return 'Low'

df_malnutrition['Malnutrition_Level'] = df_malnutrition['Mean_Estimate'].apply(get_malnutrition_level)
```
